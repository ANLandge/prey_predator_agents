```python
"""
1. Create a class `LivingThing` with attributes - `genus`, `species`.
2. Make an instance of the `LivingThing` class with attributes of your choice. Print the attributes of the instance.
"""

class LivingThing:
    def __init__(self, genus, species):
        self.genus = genus
        self.species = species

my_living_thing = LivingThing("Homo", "sapiens")
print(my_living_thing.genus)  # Output: Homo
print(my_living_thing.species)  # Output: sapiens



```

```python

"""
Create child classes of LivingThing called Bacteria, Protists, Plants, Fungi, and Animals. 
The child classes should inherit all the attributes of the parent class. Add an attribute called kingdom to each child class.

Make at least one instance of each of the child classes. Print the attributes of the instance.

"""

class Bacteria(LivingThing):
    kingdom = "Monera"

class Protists(LivingThing):
    kingdom = "Protista"

class Plants(LivingThing):
    kingdom = "Plantae"

class Fungi(LivingThing):
    kingdom = "Fungi"

class Animals(LivingThing):
    kingdom = "Animalia"

my_bacteria = Bacteria("Escherichia", "coli")
print(my_bacteria.genus)  # Output: Escherichia
print(my_bacteria.species)  # Output: coli
print(my_bacteria.kingdom)  # Output: Monera

my_protist = Protists("Paramecium", "aurelia")
print(my_protist.genus)  # Output: Paramecium
print(my_protist.species)  # Output: aurelia
print(my_protist.kingdom)  # Output: Protista

my_plant = Plants("Arabidopsis", "thaliana")
print(my_plant.genus)  # Output: Arabidopsis
print(my_plant.species)  # Output: thaliana
print(my_plant.kingdom)  # Output: Plantae

my_fungi = Fungi("Saccharomyces", "cerevisiae")
print(my_fungi.genus)  # Output: Saccharomyces
print(my_fungi.species)  # Output: cerevisiae
print(my_fungi.kingdom)  # Output: Fungi

my_animal = Animals("Homo", "sapiens")
print(my_animal.genus)  # Output: Homo
print(my_animal.species)  # Output: sapiens
print(my_animal.kingdom)  # Output: Animalia
```

```python

"""
Create a class called Vehicle that has attributes make, model, and year. 
Add a method called print_info that prints out the vehicle's make, model, and year.

Create a class called Car that inherits from Vehicle. 
Add an attribute called num_doors and a method called print_doors that prints out the number of doors the car has.

Create a class called Truck that also inherits from Vehicle. 
Add an attribute called bed_size and a method called print_bed_size that prints out the truck's bed size.

Create an instance of the Car class and call the print_info and print_doors methods.

Create an instance of the Truck class and call the print_info and print_bed_size methods.

"""

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def print_info(self):
        print(f"{self.year} {self.make} {self.model}")


class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def print_doors(self):
        print(f"This car has {self.num_doors} doors.")

class Truck(Vehicle):
    def __init__(self, make, model, year, bed_size):
        super().__init__(make, model, year)
        self.bed_size = bed_size

    def print_bed_size(self):
        print(f"This truck has a {self.bed_size}-foot bed.")

my_car = Car("Toyota", "Corolla", 2022, 4)
my_car.print_info()  # Output: 2022 Toyota Corolla
my_car.print_doors()  # Output: This car has 4 doors.

my_truck = Truck("Ford", "F-150", 2021, 6)
my_truck.print_info()  # Output: 2021 Ford F-150
my_truck.print_bed_size()  # Output: This truck has a 6-foot bed.

```

```python
# Exercise 16 
"""
Stacked Area Chart using plt.fill_between()

A stacked area chart is used to show the composition of a whole by 
showing how different parts of a data contribute to the whole over time. 
It is similar to a stacked bar chart, but is used when the data is continuous.

In this exercise, we will use the fill_between function of pyplot module in 
Matplotlib to create a stacked area chart using multiple datasets.
"""

import matplotlib.pyplot as plt
import numpy as np

# Create the data
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([3, 4, 2, 5, 1])
y3 = np.array([2, 5, 1, 4, 3])

# Create a figure and axis
fig, ax = plt.subplots()

# Stack the data and plot it as filled areas
ax.fill_between(x, y1, color='blue', alpha=0.5)
ax.fill_between(x, y1, y1+y2, color='green', alpha=0.5)
ax.fill_between(x, y1+y2, y1+y2+y3, color='red', alpha=0.5)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Stacked Area Chart')

# Show the plot
plt.show()
```

Box Plot using plt.boxplot()

A box plot is used to show the distribution of a dataset by displaying the range and distribution of the data. It is used to display the minimum, maximum, median, and quartiles of a dataset. It is useful for identifying outliers and comparing distributions.

In this exercise, we will use the boxplot function of pyplot module in Matplotlib to create a box plot and customize the color and style of the boxes, whiskers, and outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a figure and axis
fig, ax = plt.subplots()

# Create the box plot and customize the colors and style
bp = ax.boxplot(data, patch_artist=True, notch=True,
                capprops=dict(color='black', linewidth=2),
                whiskerprops=dict(color='black', linewidth=2),
                flierprops=dict(marker='o', markerfacecolor='red', markersize=5),
                boxprops=dict(facecolor='green', color='black', linewidth=2),
                medianprops=dict(color='black', linewidth=2))

# Add labels and title
ax.set_xticklabels(['Data 1', 'Data 2', 'Data 3'])
ax.set_xlabel('Data')
ax.set_ylabel('Value')
ax.set_title('Box Plot')

# Show the plot
plt.show()
```

**System desciption:**

The system consists of organisms. Some are prey and others are predators. Both types can reproduce, and die. Preadtors eat the prey and then the prey is dead. Predators reproduce when there is sufficient prey and other predators are present neayby for mating. They preys reproduce at a constant rate. 

 The prey-predator system can be described by the classical lotka-volterra system of equations - 
(Ref. - Shim, H. & Fishwick, P. A. doi:10.1016/B978-008045405-4.00225-1.)

 Predator = Fox = F

 Prey = Rabbit = R

 The **assumptions** are 
 1. Rabbits reproduce at a rate dependent on current rabbit population (exponential growth). 
 2. Foxes die at a rate dependent on current fox population (exponential decay/death).
 3. Rabbits are eaten by foxes at rate dependent on both rabbit and fox populations.
 4. Foxes reproduce at rate dependent on both rabbit and fox populations. 
 
 The **parameters** are - 
* a - rabbit birth rate constant
* b - rabbit death rate constant
* c - fox birth rate constant
* d - fox death rate constant


 This gives us the **equations**- 

* *dR/dt = a X R - b X F X R* 

* *dF/dt = c X F X R - d X F*

Aims:
1. To simulate the spatiotemporal dynamics of the system

2. To visualize the pre-predator interactions and population change over time

3. To Test model behaviour at different parameters, initial conditions, etc. 

4. To have some fun



```python
"""
This program provides classes and methods to simulate a simple agent-based
predator-prey model.

"""
# import the necessary packages

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime
import time
import os
from IPython import display

%matplotlib inline 
# use matplotlib inline mode to disply plots

class organism:
    """
    A general class that contains all the agents.
    """
    def __init__(self, id, typ, sz, rng):
        self.typ= typ # 'prey' or 'predator'
        self.id=str(id).zfill(4)  # each agent gets an id
        self.name=typ+self.id # and a name
        a=np.random.randint(0,sz) # a random x-coordinate
        b=np.random.randint(0,sz) # a random y-coordinate
        self.pos=np.array([a,b]) # initial position of the agent
        self.rng = rng # the range of the agent


class simulateModel:
    """
    This class contains methods to update the system state
    """
    def __init__(self, steps, outdir, landSize, initial_prey, initial_predator, rngPre, rngPrd, a, b, c, d):
        self.steps=steps 
        self.sz = landSize # the size of the landscape where the organisms live
        self.outdir= outdir
        self.preyCnt = initial_prey
        self.prdCnt = initial_predator
        self.rngPre = rngPre 
        self.rngPrd= rngPrd
        self.a = a # prey birth rate
        self.b = b # prey death rate
        self.c = c # predator birth rate
        self.d = d # predator death rate
        self.cntDict={} # a dictionary to store prey and predator counts
        self.cntDict['prey']= [self.preyCnt]
        self.cntDict['predator']= [self.prdCnt]

    def initialize(self, ):
        """
        This function initializes the model with given number of preys and predators and adds them to a list. 
        """
        self.orgList=[]
        for i in range(self.preyCnt):
            ag= organism(i,'prey', self.sz, self.rngPre)
            self.orgList.append(ag)

        for i in range(self.prdCnt):
            ag = organism(i, 'predator', self.sz, self.rngPrd)
            self.orgList.append(ag)

    
    def moveAgent(self, ag):
        x = (ag.pos[0]+ np.random.randint(-ag.rng,ag.rng)) % self.sz
        y = (ag.pos[1]+ np.random.randint(-ag.rng,ag.rng)) % self.sz
        newPos= np.array([x,y])

        ag.pos = newPos
        pass
    
    def getNeighbours(self,ag):
        """
        This function checks who is in the vicinity of the given agent. 
        output:
        [1,1] if prey and predator is nearby
        [1,0] only prey is near
        [0,1] only predator is near
        [0,0] nobody is closeby
        """
        self.neighbours = [0,0]

        for agN in self.orgList:
            if agN.name != ag.name: # to look only for agents other than the given agent
                x1,y1 = ag.pos
                x2,y2 = agN.pos
                dist = np.sqrt((x2-x1)**2 + (y2-y1)**2) # compute euclidean distance
                if dist<=ag.rng:
                    if agN.typ=='prey':
                        self.neighbours[0] = 1
                    elif agN.typ=='predator':
                        self.neighbours[1] = 1
            if self.neighbours==[1,1]:
                break

    def update(self,step):
        """
        The logic - at each step
        1. the prey reproduces with probability 'a' 
        2. the prey dies with probability 'b' (if a predator is nearby)
        3. the predator reproduce with probability 'c' (if a prey is nearby)
        4. the predator dies at probability 'd' 

        """
        for ag in self.orgList:
            self.moveAgent(ag)
            self.getNeighbours(ag) # first see who is nearby

            if ag.typ=='prey':  # the logic for prey
                # if self.neighbours[0]==1:
                self.doIt = np.random.choice([0,1], p=[1-self.a, self.a]) # applying the probability
                if self.doIt:
                    # add new prey
                    agN = organism(self.preyCnt,'prey', self.sz, self.rngPre)
                    agN.pos= ag.pos
                    self.orgList.append(agN) 
                    self.preyCnt+=1 

                if self.neighbours[1]==1:
                    self.doIt = np.random.choice([0,1], p=[1-self.b, self.b])
                    if self.doIt:
                        # the prey gets eaten 
                        self.orgList.remove(ag) 
                        self.preyCnt-=1 

            elif ag.typ=='predator':
                if self.neighbours[0]==1:
                    self.doIt = np.random.choice([0,1], p=[1-self.c, self.c])
                    if self.doIt:
                        # the predator reproduces
                        agN = organism(self.prdCnt,'predator', self.sz, self.rngPrd)
                        agN.pos= ag.pos
                        self.orgList.append(agN) 
                        self.prdCnt+=1 

                if self.neighbours==[0,1]:
                    self.doIt = np.random.choice([0,1], p=[1-self.d, self.d])
                    if self.doIt:
                        # the predator dies
                        self.orgList.remove(ag) # remove organism from the list
                        self.prdCnt-=1 #decrease prdCnt by 1

        #update prey and predator counts
        self.cntDict['prey'].append(self.preyCnt)
        self.cntDict['predator'].append(self.prdCnt)

    def plotAgents(self, step):
        """
        This function makes the plots of prey and predator distribution and counts.
        """
        plt.clf()
        display.clear_output(wait=True)

        plt.subplot(1,2,1) # to show the distribution
        plt.title(f"step_{step}")
        plt.xticks([])
        plt.yticks([])
        plt.xlim(0,self.sz)
        plt.ylim(0,self.sz)
        
        for ag in self.orgList:
            if ag.typ=="prey":
                plt.plot(ag.pos[0], ag.pos[1], 'o', color='k')
            elif ag.typ=="predator":
                plt.plot(ag.pos[0], ag.pos[1], 'x', color='b')
            else:
                None
      
        plt.subplot(1,2,2) # to show the counts
        plt.xlabel("Step number")
        plt.ylabel("Counts")

        xAx = np.linspace(0,step+1,step+2)
        plt.plot(xAx, self.cntDict['prey'], '-', color='k', label='prey')
        plt.plot(xAx, self.cntDict['predator'], '-', color='b', label='predator')    
        plt.legend()
        
        display.display(plt.gcf())        
        stepID = f"{step}".zfill(4)
        outName = self.outdir/f"agentsAt_{stepID}.tif"
        plt.savefig(outName)
    


```

```python
print("\n**************************************************** \nStarting the simulation program...")
t1 = time.time()

#make directory to store the results
cwdPath=Path(os.path.abspath(os.getcwd()))

#get the current date and time to create a unique directory for each simulation
now=datetime.now()
datetime_str=(now.strftime("%Y%m%d_%H%M%S_"))

dirName=str(datetime_str+"prey_predator")
outdir=cwdPath/'data'/dirName
outdir.mkdir(mode=0o777, parents=True, exist_ok=True)

print("Created directory for results ...")

# define parameters
alpha=0.1
beta=0.4
delta=0.2
gamma=0.8

steps = 500
landSize = 800
initial_prey= 40
initial_predator = 40
rngPre = 20
rngPrd = 40

# populate the land with some organisms
print("Running simulation ...")
sim1 = simulateModel(steps, outdir, landSize, initial_prey, initial_predator, rngPre, rngPrd,
  alpha, beta, delta, gamma)

sim1.initialize()

# update the population
plt.figure(figsize= [11, 5])

for step in range(steps):
    sim1.update(step)
    sim1.plotAgents(step)

    if sim1.preyCnt==0:
        print("All preys died.")
        break

    if sim1.prdCnt==0:
        print("All predators died.")
        break

    if sim1.preyCnt>2000:
        print(f"Maximum prey capacity reached. prey count = {sim1.preyCnt}")
        break

    if sim1.prdCnt>2000:
        print(f"Maximum predator capacity reached. predator count = {sim1.prdCnt}")
        break
plt.close()

t2= time.time()
totalSec= t2-t1
Sec=int(totalSec%60)
Hrs=int(totalSec//3600)
Min=int((totalSec%3600)//60)

print ("Program completed in %sHr:%sMin:%ssec\n"%(Hrs,Min,Sec))
```
